```python
wgmma_M = 64 
wgmma_N = 128
wgmma_K = 16 



A_smem_cost_in_bytes = wgmma_M*wgmma_K*2 
B_smem_cost_in_bytes = wgmma_K*wgmma_N*2 
print(A_smem_cost_in_bytes + B_smem_cost_in_bytes)
c_frag_cost_in_bytes = wgmma_M*wgmma_N*4
print(c_frag_cost_in_bytes)

buffers_per_fragment = c_frag_cost_in_bytes/(A_smem_cost_in_bytes + B_smem_cost_in_bytes)
print(buffers_per_fragment)
N_fragments = 20 
N_buffers = N_fragments*2 


def cost_per_SM(wgmma_M, wgmma_K, wgmma_N, N_fragments, N_buffers_per_fragment): 
  N_buffers = N_fragments*N_buffers_per_fragment
  A_bytes = wgmma_K*wgmma_M*2 
  B_bytes = wgmma_K*wgmma_N*2 
  c_frag_bytes = wgmma_N*wgmma_M*4 
  smem_cost = (A_bytes + B_bytes)*N_buffers
  register_cost = c_frag_bytes*N_fragments
  return smem_cost/(1000), register_cost/(1000)
  
  
  
```

```python
smem_cost, register_cost = cost_per_SM(wgmma_M, wgmma_K, wgmma_N, 8,4)
```
