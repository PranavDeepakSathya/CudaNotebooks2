```python

class all_token: 
    def __repr__(self): return "ALL"

# Create the singleton instance
ALL = all_token()

# 2. The Coordinate Class
class Coord: 
    def __init__(self, coordinates: Tuple[Union[int, all_token], ...]): 
        # Convert list to tuple if necessary for immutability
        if isinstance(coordinates, list):
            coordinates = tuple(coordinates)
            
        # --- Basic Asserts ---
        assert isinstance(coordinates, tuple), \
            f"Expected tuple, got {type(coordinates)}"
            
        for i, x in enumerate(coordinates):
            # We allow int or the specific ALL instance
            assert isinstance(x, int) or isinstance(x, all_token), \
                f"Invalid type at index {i}: {type(x)}. Must be int or ALL."
        
        self.val = coordinates
        self.len = len(coordinates)
```
